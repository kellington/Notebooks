## Picks

```python
picks_page = 'http://www.espn.com/nfl/picks'
```

```python
# Columns based on Analyst name and a fix
all_cols = df_picks.iloc[1]
all_cols[0] = 'GameInfo'
```

```python
df_picks.columns = all_cols
df_picks
```

```python
# just the named lick cols
df_picks_cols = df_picks.columns
df_picks_cols = df_picks_cols[1:-3]
df_picks_cols
```

```python
#df_picks[0].str.contains("@")
df_picks_games = df_picks[df_picks['GameInfo'].str.contains("@", na=False)].copy()
```

```python
#df_picks_games.reset_index(drop = True)
df_picks_games.index = np.arange(1, len(df_picks_games) + 1)
df_picks_games
```

```python
#reOrganize
df_picks_games.index.name = 'Game'
df_banana = df_picks_games.rename_axis(None, axis="columns")
```

```python
df_banana
```

```python
def summarize_predictions (row):
    row_sum = Counter(row)
    return row_sum
```

```python
df_banana['Predictions'] = df_banana[df_picks_cols].apply(summarize_predictions,axis=1)
```

```python
#note, the object column is still a counter
type(df_banana.iloc[2]['Predictions'])
```

```python
def most_picked(prediction_counter):
    fav = prediction_counter.most_common(1)
    #print(prediction_counter)
    #print(fav)
    #print(type(fav[0]))
    #print(fav[0][0])
    f = fav[0][0]
    
    return f

df_banana['Favorite'] = df_banana['Predictions'].apply(most_picked)
```

```python
df_apple = df_banana[['GameInfo','Favorite','Predictions']].copy()
```

```python
df_apple
```

```python
def parse_gameinfo(gameinfo_string):
    # split and then use ~deliminator
    pieces = gameinfo_string.split()
    vis = pieces[0]
    home =pieces[2][:-3]
    day = pieces[2][-3:]
    time = pieces[3]    
    combined = vis + '-' + home + '-' + day + '-' + time
    return combined
```

```python
df_apple['parsed'] = df_apple['GameInfo'].apply(parse_gameinfo)
```

```python
df_apple[['Visitor','Home','Day','Time']] = df_apple.parsed.str.split("-",expand=True,)
```

```python
df_apple
```

```python
df_apple.to_csv('picks.csv')
```
